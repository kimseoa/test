```
import pandas as pd
train_data= pd.read_table('/content/gdrive/My Drive/mountfile/ratings_train1.txt')
test_data= pd.read_table('/content/gdrive/My Drive/mountfile/ratings_test1.txt')
```

```
#null 제거
train_data=train_data.dropna(how='any') # Null 값이 존재하는 행 제거
print(train_data.isnull().values.any()) # Null 값이 존재하는지 확인
```

```
# 한글과 공백을 제외하고 모두 제거
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
train_data[:5]
```

```
#불용어 정의
stopwords=['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```
#토큰화 및 불용어 제거

import konlpy
from konlpy.tag import Okt

#한국어 
okt = Okt()  # 형태소 분석
X_train=[]
for sentence in train_data['document']:
    temp_X = []
    temp_X=okt.morphs(sentence, stem=True) # 토큰화
    temp_X=[word for word in temp_X if not word in stopwords] # 불용어 제거
    X_train.append(temp_X)
```

```
print(X_train[:3])
```

```
#라벨

X_train_label=[]
for sentence in train_data['label']:
    X_train_label.append(sentence)
```

```
len(X_train)
```

```
len(X_train_label)
```

```
#라벨, 리스트 묶기

train_data_a=[]
for i in range(len(X_train)):
  dict = { 'text' : 1, 'label' : 2 }
  dict['text'] = X_train[i]
  dict['label']= str(X_train_label[i])
  train_data_a.append(dict)
train_data_a
```

```
#test 내용 만들고

#한국어 
okt = Okt()  # 형태소 분석
X_test=[]
for sentence in test_data['document']:
    temp_X = []
    temp_X=okt.morphs(sentence, stem=True) # 토큰화
    temp_X=[word for word in temp_X if not word in stopwords] # 불용어 제거
    X_test.append(temp_X)
```

```
#test라벨

X_test_label=[]
for sentence in test_data['label']:
    X_test_label.append(sentence)
```

```
#테스트 라벨, 리스트 묶기

test_data_a=[]
for i in range(len(X_test)):
  dict = { 'text' : 1, 'label' : 2 }
  dict['text'] = X_test[i]
  dict['label']= str(X_test_label[i])
  test_data_a.append(dict)
test_data_a
```
